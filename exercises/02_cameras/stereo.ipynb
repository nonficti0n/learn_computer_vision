```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import sys
%matplotlib inline
plt.rcParams['figure.figsize'] = [12, 8]
```

```python
# Load a color image in grayscale
img_l = cv2.imread('scene1.row3.col1.ppm',0)
img_r = cv2.imread('scene1.row3.col2.ppm',0)

#img = cv2.imread('boat.jpg',0)

plt.imshow(img_l,cmap = 'gray')
plt.title('Greyscale image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
stereo = cv2.StereoBM_create(numDisparities=16, blockSize=9)
disparity = stereo.compute(img_l,img_r)
plt.imshow(disparity,'gray')
plt.show()
```
